## Import the Dataset

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'
df = pd.read_csv(url)

df.shape
df.columns
df.head(3)
df.tail(3)
```

## Slice the dataframe from 'school' until the 'guardian' column

```python
stud_alcoh = df.loc[:, 'school':'guardian']

stud_alcoh.shape
stud_alcoh.columns
stud_alcoh.head(3)
stud_alcoh.tail(3)
```

## <font color=red>lambda function</font>

> apply()

> applymap()

### apply

```python
capitalizer = lambda x: x.capitalize()
```

```python
stud_alcoh['Mjob'] = stud_alcoh['Mjob'].apply(capitalizer)
stud_alcoh['Fjob'] = stud_alcoh['Fjob'].apply(capitalizer)

stud_alcoh[['Mjob', 'Fjob']].head(3)
```

```python
def majority(age):
    if age > 17:
        return True
    else:
        return False
```

```python
stud_alcoh['legal_drinker'] = stud_alcoh['age'].apply(majority)
stud_alcoh.head(3)
```

### applymap

```python
def times10(x):
    if type(x) is int:
        return x * 10
    return x
```

```python
stud_alcoh.applymap(times10).head()
```
