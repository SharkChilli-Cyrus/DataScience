**Import Libraries**

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

## Import the Dataset

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv'

crime = pd.read_csv(url)
crime.shape
crime.columns
crime.head(3)
crime.tail(3)
```

## The type of the columns

```python
crime['Year'] = pd.to_datetime(crime.Year, format='%Y')
crime.info()
```

```python
crime.head(3)

crime.set_index('Year', drop=True, inplace=True)

crime.head(3)
```

##  Delete the Total column

```python
del crime['Total']
crime.head(3)
```

## Group the year by decades and sum the values

**NOTE: Pay attention to the population column number, summing this column is a mistake**

```python
# Sum each decade
crimes = crime.resample('10AS').sum()

# Get the max value only for the 'Population' column
population = crime['Population'].resample('10AS').max()

crimes['Population'] = population
crimes.head(3)
```

```python
crimes.index.name = None
crimes.index = crimes.index.year

crimes.head()
```

## What is the mos dangerous decade to live in the US?

```python
crimes.idxmax(0)
```
