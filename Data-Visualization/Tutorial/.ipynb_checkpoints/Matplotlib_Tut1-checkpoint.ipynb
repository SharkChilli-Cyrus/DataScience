**NOTE:** The difference between Axes and Axis<br>
<u>The **Axes** contains two (or three in the case of 3D) **Axis** objects (<font color=red>be aware of the difference between Axes and Axis</font>)</u>

### Axes

This is a plot, it is the region of the image with the data space.<br>
**A given figure can contain many Axes, but a given `Axes` object can only be in one `Figure`.**

<u>The **Axes** contains two (or three in the case of 3D) **Axis** objects (<font color=red>be aware of the difference between Axes and Axis</font>)</u>
* **Axes:**<br>
Each Axes has a **title** (set via `set_title()`), an **x-label** (set via `set_xlabel()`), and a **y-label** set via `set_ylabel()`).
* **Axis:**<br>
Can **take care of the data limits** (the data limits can also be controlled via set via the `set_xlim()` and `set_ylim()` Axes methods).

### Axis

These are the number-line-like objects. They take care of setting the graph limits and generating the **<font color=red>ticks</font>** (the marks on the axis) and **ticklabels** (strings labeling the ticks).

The location of the ticks is determined by a **Locator** object and the ticklabel strings are formatted by a **Formatter**. The combination of the correct Locator and Formatter gives very fine control over the tick locations and labels.

## Types Of Inputs To Plotting Functions

All of plotting functions expect np.array or np.ma.masked_array as input. Classes that are 'array-like' such as pandas data objects and np.matrix may or may not work as intended. It is best to convert these to np.array objects prior to plotting.

```python
a = pd.DataFrame(np.random.rand(3,4),
                 columns = ['a', 'b', 'c', 'd'])
a_asarray = a.values
a_asarray
```

```python
b = np.matrix([[1,2],
               [3,4]])
b_asarray = np.asarray(b)
b_asarray
```

## Examples

```python
'''
Example 1
'''

x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear') # creates the axes

# add additional lines on the same axes
plt.plot(x, x**2, label='quadratic') 
plt.plot(x, x**3, label='cubic')

# set the axes labels
plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot") # Set the title

plt.legend() # add a legend

plt.show()
```

```python
x1 = np.arange(0, 10, 0.2)
y1 = np.sin(x1)
x2 = np.arange(0, 10, 0.2)
y2 = np.cos(x2)
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x1, y1)
ax2.plot(x2, y2)
plt.show()
```

```python
def my_plotter(ax, data1, data2, param_dict):
    '''
    A helper function to make graph
    
    Parameters
    ----------
    ax: Axes
        The axes to draw to
    
    data1: array
        The x data
        
     data2: array
         The y data
     
     param_dict: dict
         Dictionary of kwargs to pass to ax.plot
     
     Returns
     -------
     out: list
         list of artists added
    '''
    out = ax.plot(data1, data2, **param_dict)
    return out

# which you would then use as:

data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})
plt.show()
```

Have 2 sub-plots

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
plt.show()
```
