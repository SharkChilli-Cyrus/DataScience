```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3,50) # 用np.linspace定义x,范围是(-3,3),个数是50
y1 = 2*x+1
y2 = x**2

plt.figure() # 定义一个图像窗口用于画x,y1的图像
plt.plot(x,y1) # 第一个位置是x轴参数，第二个位置是y轴参数
plt.show()

plt.figure(num=3, figsize=(8,5)) # 定义的图像窗口编号为3，大小为(8,5)
plt.plot(x, y1)
plt.plot(x, y2, color='red', linewidth=1.0, linestyle='--') 
#颜色属性(color)为红色(red)，曲线宽度(linewidth)为1.0，曲线类型(linestyle)为虚线(--)
plt.show()
```

```python
### set axis 1 ###
#### 设置坐标轴范围，名称，刻度，对应刻度名称 ####

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2

plt.figure()
plt.plot(x, y1)
plt.plot(x, y2, color='red', linewidth=1.0, linestyle='--') 

plt.xlim(-1,2) # 设置x轴范围(-1,2)
plt.ylim(-2,3) # 设置y轴范围(-2,3)

plt.xlabel('I am x') # 设置x轴名称
plt.ylabel('I am y') # 设置y轴名称

new_ticks = np.linspace(-1,2,5) # 定义范围和个数
print(new_ticks)

plt.xticks(new_ticks) # 设置x轴刻度
plt.yticks(
    [-2, -1.8, -1, 1.22, 3],
    [r'$really\ bad$', r'$bad\ \alpha$', r'$normal$', r'good', r'$really\ good$'])
# 设置y轴刻度以及对应名称，$ $内表示数学写法 , 用\来转置出空格

plt.show()
```

```python
### Set Axis 2###
#### 设置坐标轴，隐藏部分图框，设置x,y坐标轴位置 ####

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 50)
y1 = 2*x+1
y2 = x**2

plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')

plt.xlim((-1,2))
plt.ylim((-2,3))

new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
plt.yticks(
    [-2, -1.8, -1, 1.5, 3], 
    ['$really\ bad$', '$bad$', '$normal$', '$good$', '$really\ good$'])

ax = plt.gca() # 获取当前坐标轴信息
ax.spines['right'].set_color('none') # .spines设置边框，右侧边框, .set_color设置边框颜色，默认白色
ax.spines['top'].set_color('none') # .spines设置边框，上边框, .set_color设置边框颜色，默认白色

ax = plt.gca() # 获取当前坐标轴信息
ax.xaxis.set_ticks_position('bottom')
# .xaxis.set_ticks_position设置x轴刻度数字或名称位置: bottom (所有位置: top, bottom, both, default, none)

ax.spines['bottom'].set_position(('data', 0))
# .spines设置边框[x轴]， .set_position设置边框位置(y=0的位置) , 位置的所有属性：outward, axes, data 

ax.spines['left'].set_position(('data', 0)) # 设置边框[y轴]

plt.show()
```

```python
### Annotation ###

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y = 2*x+1
  
plt.figure(num=1, figsize=(8,5), )
plt.plot(x, y,)

# 移动坐标轴的位置 #
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# 画出图中特定点 #
x0 = 1
y0 = 2*x0 + 1
plt.scatter(x0, y0, s=50, color='r') # set dot styles
plt.plot([x0, x0], [y0, 0], 'k--', lw=1.5) # 画出垂直于x轴的虚线

#### method 1 ####
plt.annotate(r'$2x+1=%s$' % y0, 
             xy=(x0,y0), xycoords='data', 
             xytext=(+30, -30),  textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=0.2'))
# xycoords='data' 基于数据的值来选择位置
# xytext=(+30, -30) 和 textcoords='offset points' 对于标注位置的描述和xy偏差值
# fontsize 设置文本字号
# arrowprops 对图中箭头类型的设置


#### method 2 ####
plt.text(-3.7, 3, r'$This\ is\ the\ some\ text.\ \mu\ \sigma_i\ \alpha_t$',
        fontdict={'size':15, 'color':'blue'})
# -3.7, 3 是选择text的位置
# 空格需要用转字符\ 表示
# fontdict 设置文本字体

plt.show()
```

```python
### Bar 柱状图 ###

import matplotlib.pyplot as plt
import numpy as np

n = 12
X = np.arange(n) # X 为 0-11 的整数
Y1 = (1-X/float(n))*np.random.uniform(0.5, 1.0, n)
Y2 = (1-X/float(n))*np.random.uniform(0.5, 1.0, n)


plt.figure()
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')
# facecolor 设置主体颜色， edgecolor设置边框颜色

for x, y in zip(X, Y1):
    plt.text(x+0.1, y+0.05, '%.2f' % y, ha='center', va='bottom')
# '%0.2f' 保留两位小数
# ha : horizontal alignment, 横向居中对齐
# va='bottom' 纵向底部对齐

for x, y in zip(X, Y2): # zip 是指将数值分别传入
    plt.text(x+0.1, -y-0.05, '%.2f' % y, ha='center', va='top')
# 坐标轴下方显示bar，y的数值需要加负号
# '%0.2f' 保留两位小数
# ha : horizontal alignment, 横向居中对齐
# va='top' 纵向顶部对齐


# 设置坐标轴范围 #
plt.xlim(-0.5, n)
plt.ylim(-1.25, 1.25)

# 取消坐标轴显示 #
plt.xticks(())
plt.yticks(())

plt.show()
```

```python
### Contour 等高线图 ###

import matplotlib.pyplot as plt
import numpy as np

# the height function #
def f(x,y): 
    return (1 - x/2 + x**5 + y**3) * np.exp(-x**2 - y**2)

# 数据集由x, y 以及对应的高度值组成，高度值使用定义的f(x,y)生成 #
# x, y 分别是在区间[-3, 3]中均匀分布的256个值，
# 并用meshgrid在二维平面中将每一个x和每一个y分别对应起来，编织成栅格
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

# 颜色填充 #
plt.contourf(X, Y, f(X, Y), 8, alpha=0.75, cmap=plt.cm.hot)
# 位置参数 X, Y, f(X,Y), 等高线高度被分成8+2部分，透明度0.75
# 将f(X,Y) 的值对应到 color map的暖色组寻找对应颜色

# use plt.contour to add contour lines #
C = plt.contour(X, Y, f(X, Y), 8, colors='black')
# 8代表等高线密集程度，这里被分成8+2个部分

# 添加高度数字 #
plt.clabel(C, inline=True, fontsize=10) # adding label
# inline 控制是否将label画在线里面
# fontsize 字体大小

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
### 3D 图像 ##

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y) # x-y平面的网格
R = np.sqrt(X**2 + Y**2)

Z = np.sin(R) # height value

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow'))
# rstride 代表row 的跨度
# cstride 代表column 的跨度

# 平面的投影 #
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'))
ax.set_zlim(-2,2)
# zdir 选择从哪个轴压缩成平面图，选择了 z 那么效果会是xy平面的投影

plt.show()
```

```python
### Subplot 分格显示 ###

import matplotlib.pyplot as plt
import numpy as np

##### Method 1 #####
plt.figure(num=1)

ax1 = plt.subplot2grid((3,3), (0,0), colspan=3, rowspan= 1)
# (3, 3) 表示整个图像窗口分成3行3列 (0,1,2 行；0,1,2 列)
# (0, 0) 表示从第1行第1列开始作图
# colspan=3 表示列的跨度为3 (默认值为1)
# rowspan=1 表示行的跨度为1 (默认值为1)

ax1.plot([1,2], [2,3]) # 画小图
ax1.set_title('1_title') # 设置小图标题

ax2 = plt.subplot2grid((3,3), (1,0), colspan=2) # 第2行第1列，跨2列
ax2.set_title('2_title')

ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2) # 第2行第3列，跨2行
ax3.set_title('3_title')

ax4 = plt.subplot2grid((3,3), (2,0)) # 第3行第1列
ax4.set_title('4_title')

ax5 = plt.subplot2grid((3,3), (2,1)) # 第3行第2列
ax5.set_title('5_title')

plt.show()


##### Method 2 (matplotlib.gridspec) #####
import matplotlib.gridspec as gridspec

plt.figure(num=2)

gs = gridspec.GridSpec(3,3)

ax1 = plt.subplot(gs[0, :]) # 小图占第1行的所有列
ax1.plot([2,5], [3,6])

ax2 = plt.subplot(gs[1, :2]) # (1, )第2行, ( :2)第3列前的所有列, 即第1列(0),第2列(1)
ax3 = plt.subplot(gs[1:, 2]) # 第2列, 第2行之后的所有行, 即第2行(1), 第3行(2)
# Remark:  --  :2 是2之前，不含2,  --  1:  是从1开始算，含1

ax4 = plt.subplot(gs[2, 0]) # 第3行，第1列
ax5 = plt.subplot(gs[-1, -2]) # 倒数第1行，倒数第2列

plt.show()


##### Method 3 #####

plt.figure(num=3)

f,((ax11, ax12), (ax21, ax22)) = plt.subplots(2, 2, sharex=True, sharey=True)
# 创建2行2列的图像窗口
# 第1行从左至右依次放置ax11, ax12, 第2行从左至右依次放置ax21, ax22
# sharex=True 共享x轴坐标
# sharey=True 共享y轴坐标

ax11.scatter([0, 1], [2, 3]) # 创建一个散点图

plt.tight_layout() # 紧凑显示图像
plt.show()
```

```python
### Animation 动画 ###

import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

fig, ax = plt.subplots()

# 0 ~ 2\pi 的正弦函数图像
x = np.arange(0, 2*np.pi, 0.01) # 0 -- 2\pi
line, = ax.plot(x, np.sin(x)) # sin的函数图像

# line 是个列表,  选择第1位 line,

# 构造自定义动画函数animate 用来更新每一帧上各个x对应的y, 参数i 表示第i 帧 #
def animate(i):
    line.set_ydata(np.sin(x+i/100))
    return line,

# 构造开始帧函数init #
def init():
    line.set_ydata(np.sin(x))
    return line,

ani = animation.FuncAnimation(fig=fig, 
                              func=animate, 
                              frames=100, 
                              init_func=init, 
                              interval=60, 
                              blit=False) 
# fig 进行动画绘制的figure
# func 自定义动画函数
# frames 动画长度，一次循环包含的帧数
# init_func 自定义开始帧
# interval 更新频率，以ms计
# blit 选择是否更新所有点，还是仅更新产生变化的点， Mac用户 blit 选择False

plt.show()

import ffmpeg
ani.save('basic_animation.mp4', fps=30, extra_args=['-vcodec', 'libx264'])
```
