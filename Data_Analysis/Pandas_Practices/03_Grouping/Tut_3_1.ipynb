**Import libraries**

```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

## Read raw data

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'

drinks = pd.read_csv(url)

drinks.shape
drinks.columns

drinks.head(3)
drinks.tail(3)
```

## Groupby

```python
group1 = drinks.groupby('continent')
group1.describe()
```

### Which continent drinks more beer on average?

```python
# beer_servings
group1.beer_servings.mean()
```

### For each continent print the statistics for wine consumption

```python
# wine_servings
group1.wine_servings.describe()
```

### Print the mean, median alcohol consumption per continent for every column

```python
# mean
group1.mean()

# median
group1.median()
```

### Print the mean, min and max values for spirit consumption

```python
group1.spirit_servings.agg(['mean', 'min', 'max'])

# or 
# group1['beer_servings'].agg([np.mean, np.min, np.max])
```

## Test

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
                     'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
            'Rank': [1, 2, 2, 3, 3, 4, 1, 1, 2, 4, 1, 2],
            'Year': [2014, 2015, 2014, 2015, 2014, 2015, 2016, 2017, 2016, 2014, 2015, 2017],
            'Points': [876, 789, 863, 673, 741, 812, 756, 788, 694, 701, 804, 690]}
df = pd.DataFrame(ipl_data)

df.shape
df.head(3)
df.tail(3)
```

### get_group

```python
grouped = df.groupby('Year')
for name, group in grouped:
    print(name)
    print(group)
    
print ('\n******\n',grouped.get_group(2017))
```

### agg(np.size)

```python
group2 = df.groupby('Team')

group2.describe()

group2.agg(np.size)
```

### transform

```python
score = lambda x: (x - x.mean()) / x.std()*10
group2.transform(score)
```

### filter

```python
length = lambda x: len(x) >= 3
group2.filter(length)
```
